```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt

%matplotlib inline
```

```python
ceiling = 31
maxim = 31
avger = 14

data = {'Ceiling':ceiling, 'Maximum':maxim, 'Average': avger}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize=(8, 6))

plt.bar(courses, values,
        width = 0.4)
plt.title('Test results')
plt.ylabel('Steps')
plt.grid(axis = 'y')
plt.show()
```

```python
datas = pd.read_csv("greedyData2.csv")
datas.head()
```

```python
stats_by_k = datas.groupby('k').mean()[['ceiling', 'steps']]
stats_by_k.plot.line()
```
