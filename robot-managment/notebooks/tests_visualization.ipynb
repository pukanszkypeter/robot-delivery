```python
import pandas as pd
import numpy as np
import sqlite3

from matplotlib import pyplot as plt

%matplotlib inline
```

```python
# BFS visualization

connection = sqlite3.connect("../data/memory.sqlite")
bfs_tests = pd.read_sql_query("SELECT * from bfs_measurement", connection)

print(bfs_tests.head())

connection.close()
```

```python
stats = bfs_tests.groupby('ceiling').mean()[['steps']]
stats.plot.bar().grid(axis = 'y')
```

```python
stats = bfs_tests.groupby('ceiling').max()[['steps']]
stats.plot.bar().grid(axis = 'y')
```

```python
# Greedy visualization

connection = sqlite3.connect("../data/memory.sqlite")
greedy_tests = pd.read_sql_query("SELECT * from greedy_measurement", connection)

print(greedy_tests.head())

connection.close()
```

```python
stats_by_k = greedy_tests.groupby('k').mean()[['ceiling', 'steps']]
stats_by_k.plot.line()
```

```python
stats_by_k = greedy_tests.groupby('k').max()[['ceiling', 'steps']]
stats_by_k.plot.line()
```
